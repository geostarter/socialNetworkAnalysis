```python
Followers_df = pd.read_csv(r"C:\Users\Nic\Documents\socialNetworkAnalysis\TwitterTrans.csv")
Followers_df
```

```python
Followers_df = Followers_df.astype('|S')
```

```python
#https://networkx.github.io/documentation/latest/reference/generated/networkx.convert_matrix.from_pandas_edgelist.html#networkx.convert_matrix.from_pandas_edgelist
edges = nx.from_pandas_edgelist(Followers_df,'A','B')
```
